```python
from langchain_core.tools import tool
from langchain.chat_models import ChatOpenAI
from datetime import datetime
import pytz

from langchain_community.tools import DuckDuckGoSearchResults
from langchain_community.utilities import DuckDuckGoSearchAPIWrapper

from youtube_search import YoutubeSearch
from langchain_community.document_loaders import YoutubeLoader
from typing import List

llm = ChatOpenAI(model = 'gpt-4o-mini')

@tool
def get_current_time(timezone: str, location: str) -> str:
    """현재 시각을 반환하는 함수."""
    try:
        tz = pytz.timezone(timezone)
        now = datetime.now(tz).strftime("%Y-%m-%d %H:%M:%S")
        result = f'The current time in {location} ({timezone}) is {now}.'
        print(result)
        return result
    except pytz.UnknownTimeZoneError:
        error_msg = f'Error: Unknown timezone "{timezone}". Please provide a valid timezone.'
        print(error_msg)
        return error_msg    
    
@tool
def get_web_search(query: str, search_period: str) -> str:
    """
    Args:
        query(str):검색어
        search_period(str):검색 기간 (예: 'past day', 'past week', 'past month', 'past year', 'all time')
    Returns:
        str:검색 결과 요약
    """
    warapper = DuckDuckGoSearchAPIWrapper(region = "kr-kr", time=search_period)

    print('Searching the web for:', query)
    print('Search period:', search_period)

    search  = DuckDuckGoSearchResults(
        api_wrapper=warapper,
        results_separator=';\n'
    )
    docs = search.invoke(query)
    return docs

@tool
def get_youtube_search(query: str) -> List:
    """유튜브를 검색한 후, 영상들의 내용을 반환하는 함수
    
    Args:
        query(str):검색어
    Returns:
        List:영상들의 내용 리스트
    """

    print('Searching YouTube for:', query)
    videos = YoutubeSearch(query, max_results=4).to_dict()
    contents = []
    for v in videos:
        vedio_url = 'https://youtube.com'+v['url_suffix']
        loader = YoutubeLoader.from_youtube_url(vedio_url, language=['ko', 'en'])
        content = loader.load()
        contents.append(content)
    return contents


tools = [get_current_time, get_web_search, get_youtube_search]
tool_dict = {
    "get_current_time": get_current_time,
    "get_web_search": get_web_search,
    "get_youtube_search": get_youtube_search
}
```
